```python
import random
import joblib
import matplotlib.pylab as pl
%matplotlib inline
import numpy as np
```

```python
source = 'training'
n_digits = 5
path = '../data-with-replacement/%s-%d.data' % (source, n_digits)
# path = 'data/%s-%d.data' % (source, n_digits)
data, labels = joblib.load(path)
data.shape, labels.shape
```

```python
def plot(index, data, labels):
    data = data[index]
    labels = labels[index]
    print labels.tolist()
    for image, label in zip(data, labels):
        image = np.reshape(image, (28, 28))
        pl.figure(figsize=(1, 1))
        pl.axis('off')
        pl.imshow(image, cmap='gray')
```

```python
index = random.randrange(len(data))
plot(index, data, labels)
```

```python
def extract_category(category, data, labels):
    data = np.squeeze(data)
    labels = np.squeeze(labels)
    mask = labels == category
    extracted = data[mask]
    return extracted
```

```python
extracted = extract_category(1, data, labels)
print len(extracted)
start, end = 10, 20
for image in extracted[start: end]:
    image = np.reshape(image, (28, 28))
    pl.figure(figsize=(1, 1))
    pl.axis('off')
    pl.imshow(image, cmap='gray')
```
