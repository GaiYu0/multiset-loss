```python
from collections import namedtuple
import matplotlib.pylab as pl
%matplotlib inline
```

```python
empty_table = lambda: {'training': [], 'validation': [], 'test': []}
tables = {
    'ce_loss': empty_table(),
    'regression_loss': empty_table(),
    'rl_loss': empty_table(),
    'semi_cross_entropy': empty_table()
}
```

```python
base_path = '../temporary'
```

```python
for key, table in tables.items():
    for i in range(1, 10):
        path = '%s/%s%d' % (base_path, key, i)
        f = open(path, 'r')
        for line in f:
            partition, accuracy = line.split()
            tables[key][partition].append(float(accuracy))
```

```python
for partition in ('training', 'validation', 'test'):
    pl.figure()
    pl.title(partition)
    pl.xlabel('number of digits')
    pl.ylabel('accuracy')
    for key, table in tables.items():
        pl.plot(range(1, len(table[partition]) + 1), table[partition], label=key)
    pl.legend(framealpha=0, loc='best')
```
